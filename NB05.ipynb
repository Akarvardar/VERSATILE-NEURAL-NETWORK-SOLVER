```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
import vnns
```

<font color=darkblue> So far in the earlier notebooks, we have used gradient descent ('GD') to find the network parameters theta. In this notebook, we will be using some of the other algorithms available under scipy.optimize.minimize, such as 'TNC', 'SLSQP','CG', 'BFGS', 'L-BFGS-B' for comparison with the custom implementation of 'GD'. Running these alternative algorithms are equally simple and require only varying the method name accordingly when creating the DeepLearnAuto() instance. __One potential caveat for scipy methods is the fact that the results may vary according to which exact machine they were run and/or under what python/scipy version.__ For  more information about these inconsistencies, see for instance: https://stackoverflow.com/questions/46043768/same-optimization-code-different-results-on-different-computers and https://software.intel.com/en-us/forums/intel-distribution-for-python/topic/721634.

<font color=darkblue> For benchmark purposes we will focus on a network with 3 hidden layers and solve the parameters for a relatively complex yet visualizable dataset (complex4). In addition, we will set polydeg = 1 and lamda = 'auto'.

```python
scoreboard = pd.DataFrame()
for method in ['GD','TNC', 'SLSQP','CG', 'BFGS', 'L-BFGS-B']:
    print("\n\n*** Method:", method)
    dla = vnns.DeepLearnAuto(datafile = "Data\\Medium\\complex4_tr.csv", 
                             normalize = True,
                             polydeg = 1,
                             method = method,                            
                             initialize = 'deep',
                             learning_rate= 'auto', 
                             lamda = 'auto',
                             maxiter = 10000, 
                             hidden_units = 'auto',
                             num_hidden_layers = 3, 
                             total_units_max = 30,  
                             plot_J = True, 
                             report_summary = True, 
                             plot_output = True, 
                             plot_test = True,                 
                             plot_hidden = False, 
                             plot_lam = True)
    summary = dla.generate_summary()
    display(summary)
    scoreboard = scoreboard.append(summary,ignore_index=True)
print ("\n\n*** Summary ***")
display(scoreboard)
```

<font color=darkblue> The number one observation here is that for comparable accuracy levels, all scipy methods, but especially 'TNC' and 'SLSQP' reduce the number of iterations substantially (> 100x!) compared to custom gradient-descent implementation. 'TNC' is usually a more reliable method than 'SLSQP' which displays a spike in the cost function trends above. Another item to notice is that scipy methods turn out use less number of hidden units on each layer (6 or 7) vs. 8 in 'GD'. And yet two of the methods ('TNC' and 'L-BFGS-B') were able to achieve higher scores than 'GD'. Finally, except again 'TNC' and 'SLSQP', same lamda of 0.3 has been auto-calculated by the script for 'GD' and remaining three scipy algorithms. Below we have one more comparison by using a fixed hidden layer and lamda ([8,8,8] and 0.3 as calculated for 'GD') for all scipy methods:

```python
scoreboard = pd.DataFrame()
for method in ['TNC', 'SLSQP','CG', 'BFGS', 'L-BFGS-B']:
    dla = vnns.DeepLearnAuto(datafile = "Data\\Medium\\complex4_tr.csv", 
                             normalize = True, 
                             polydeg = 1,
                             method = method,                             
                             initialize = 'deep',
                             lamda = 0.3,
                             maxiter = 10000, 
                             hidden_units = [8,8,8])
    summary = dla.generate_summary()
    scoreboard = scoreboard.append(summary,ignore_index=True)
print ("\n\n*** Summary (polydeg = 1, hidden_units = [8,8,8], lamda = 0.3) ***")
display(scoreboard)
```

<font color=darkblue> What we now notice is that although the number of iterations have usually increased slightly vs. the prior run with lesser number of hidden units, they are still substantially lower for scipy methods compared to custom 'GD'. Also, hidden layer configuration and lamda being the same for all methods, four out of five scipy methods now achieve higher scores than 'GD'.

<font color=darkblue> __In conclusion, scipy.optimize.methods typically are significantly more effective than basic gradient descent implementations as long as they do not suffer from non-obvious machine-to-machine and python/scipy version-to-version differences.__
