## <font color=darkblue> NETWORK WITH ONE HIDDEN LAYER

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
import vnns
```

<font color=darkblue>In this notebook, we will be identifying the datasets that can be modeled using a neural network with a single hidden layer (which would make the network a three-layer net including input and output layers.) We will again start here by displaying the decision boundaries and summary tables for the remaining 11 datasets for which a linear decision boundary was insufficient to separate different classes. Compared to logistic regression in the previous notebook, we set hidden_units = 'auto',  num_hidden_layers = 1 (single hidden layer)  and total_units_max = 10,  which means the number of hidden units in the hidden layer cannot exceed 10 (which is an arbitrary selection here.) As such, the candidate hidden layers are: [ ] (always include the reference logistic regression case with no hidden layers) and then [number of output units + 1], [number of output units + 2].. [10]. 

```python
for data in ['elliptic1','elliptic2','elliptic3','elliptic4','complex1','complex2','complex3','complex4','complex5',
             'multiclass2','multiclass3']:
    print("\n\n*** Data:", data, "***")
    dla = vnns.DeepLearnAuto(datafile = "Data\\Medium\\"+data+"_tr.csv", 
                             normalize = True,
                             polydeg = 1,
                             method = 'GD',                             
                             initialize = 'deep',
                             learning_rate= 'auto', 
                             lamda = 'auto',
                             maxiter = 10000, 
                             hidden_units = 'auto',
                             num_hidden_layers = 1, 
                             total_units_max = 10,
                             plot_output = True)
    summary = dla.generate_summary()
    display(summary)
```

<font color=darkblue> It turns out a network with a single hidden layer does at least a decent job on all remaining datasets as can be confirmed from the training scores that are all above 0.75. Again, based on our criterion of training score + test score >= 1.8, a network with a single hidden layer seems adequate for five of the 11 datasets:  elliptic3, elliptic4, complex1, complex5, multiclass2. Below, we run once again those files by turning the informative flags on. In addition, we set plot_hidden = True in order to display the scores for all candidate hidden units. You can refer to class DeepLearnAuto() and its find_best_poly_or_hu method to understand how exactly the optimal number of hidden units is automatically computed.

```python
scoreboard = pd.DataFrame()
for data in ['elliptic3','elliptic4','complex1','complex5','multiclass2']:
    print("\n\n*** Data:", data, "***")
    dla = vnns.DeepLearnAuto(datafile = "Data\\Medium\\"+data+"_tr.csv", 
                             normalize = True,
                             polydeg = 1,
                             method = 'GD',                             
                             initialize = 'deep',
                             learning_rate= 'auto', 
                             lamda = 'auto',
                             maxiter = 10000, 
                             hidden_units = 'auto',
                             num_hidden_layers = 1, 
                             total_units_max = 10, 
                             plot_J = True, 
                             report_summary = True, 
                             plot_output = True,
                             plot_test = True,
                             plot_hidden = True,
                             plot_lam = True)
    summary = dla.generate_summary()
    scoreboard = scoreboard.append(summary,ignore_index=True)
print ("\n\n*** Summary ***")
display(scoreboard)
```

<font color=darkblue> Overall lamda remains small in most examples since regularization has been partially embedded in the automated calculation of number of hidden units. Note that rhe optimal number of hidden units varies between 3 and 5 for all five sets of data although the maximum allowed number was 10.
